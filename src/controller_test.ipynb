# Part 2: Get labels from filename

```python
def get_vels_from_filename(filename: str):
    return filename.split(']')[0].split('[')[1].split(',')


def get_action_from_vels(vels: list):
    linear_vel, angular_vel = [float(v) for v in vels]
    if angular_vel == 0: # forward
        action = 0 
    else:
        if angular_vel > 0: 
            action = 1 # left
        else: action = 2 # right
    return action
```

```python
get_action_from_vels(get_vels_from_filename(frame0))
```

```python
for im in os.listdir('../assets/images/outer_lap/'):
    print('----')
    print(im)
    a = get_action_from_vels(get_vels_from_filename(im))
    print(a)
    print(tf.one_hot(a, 3))
```

```python
# === Show hsv in channel order (R,G,B) => (H,S,V)
# note: had to reverse channel order since cv2 works with GBR
grass_hsv = cv2.cvtColor(grass_img, cv2.COLOR_BGR2HSV)
cv2.imshow('h,s,v', grass_hsv[:,:,::-1])
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# === Dark grass image to tune HSV thresholds
IMGPATH = '../assets/images/outer_lap/[0.0, 1.0]2023-03-18 19:33:33.120816.png'
grass_hsv = cv2.cvtColor(grass_img, cv2.COLOR_BGR2HSV)
grass_img = cv2.imread(IMGPATH)
grass_min = np.asarray([0,40,0])
grass_max = np.asarray([255,60,255])
grass_mask = grass_hsv[:,:,0].copy()
grass_mask = cv2.inRange(grass_img, grass_min, grass_max, grass_mask)
cv2.imshow('mask grass road', grass_mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
