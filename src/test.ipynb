```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
PATH = '/home/fizzer/Insync/1mackenziekyle@gmail.com/Google Drive/enph353/enph353_competition/assets/images/outer_lap/color_pedestrian_walkway1/'
walkway_images = os.listdir(PATH)
walkway_images[:3]
red_image = cv2.imread(PATH + walkway_images[40])
```

```python
plt.imshow(red_image[:,:,::-1])
```

```python
cv2.imshow('walkway', red_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hsv_red_image = cv2.cvtColor(red_image, cv2.COLOR_BGR2HSV)

cv2.imshow('hsv walkway', hsv_red_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_hsv = cv2.cvtColor(red_image, cv2.COLOR_BGR2HSV)


min_red = np.asarray([0, 200, 200])   # lower end of blue
max_red = np.asarray([255, 255, 255])   # upper end of blue
# Binary mask for blue pixels
mask = cv2.inRange(img_hsv, min_red, max_red)
plt.imshow(mask)
```
