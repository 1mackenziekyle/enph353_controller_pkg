```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
PATH = '/home/fizzer/Insync/1mackenziekyle@gmail.com/Google Drive/enph353/enph353_competition/assets/images/outer_lap/color_pedestrian_walkway1/'
walkway_images = os.listdir(PATH)
walkway_images[:3]
red_image = cv2.imread(PATH + walkway_images[40])
```

```python
plt.imshow(red_image[:,:,::-1])
print(red_image.shape, cv2.cvtColor(red_image, cv2.COLOR_BGR2GRAY).shape)
```

```python
cv2.imshow('walkway', red_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hsv_red_image = cv2.cvtColor(red_image, cv2.COLOR_BGR2HSV)

cv2.imshow('hsv walkway', hsv_red_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_hsv = cv2.cvtColor(red_image, cv2.COLOR_BGR2HSV)


min_red = (0, 200, 200)   # lower end of blue
max_red = (255, 255, 255)   # upper end of blue

min_shirt =(5,60,60)
max_shirt = (12,90,90)
# Binary mask for blue pixels
mask = cv2.inRange(img_hsv, min_red, max_red) + cv2.inRange(img_hsv, min_shirt, max_shirt)
cv2.imshow('mask', mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Make a random plot...
fig = plt.figure()
fig.add_subplot(111)
plt.bar(['F', 'L', 'R'], [0.5,0.2,0.3])

# If we haven't already shown or saved the plot, then we need to
# draw the figure first...
fig.canvas.draw()

# Now we can save it to a numpy array.
data = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
data = data.reshape(fig.canvas.get_width_height()[::-1] + (3,))

cv2.imshow('data', data)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
